# Generated datasets

Datasets created with simple shapes and patters that are used to evaluate the representation of the method and the different projections.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation
import json
import pandas as pd

plt.style.use("ggplot")
```

```python
#creating the events
points = []
for i in range(4):
    for j in range(4):
        x0 = 1 + 4 * i
        y0 = 1 + 4 * j
        x1 = x0 + 2
        y1 = y0 + 6
        points += [[x0, y0],
                 [x1, y0],
                 [x1, y1],
                 [x0, y1]]

points = np.array(points).astype(np.float32)
for i in range(2):
    points[:, i] = points[:, i] - np.min(points[:, i])
    points[:, i] = points[:, i] / np.max(points[:, i])
    points[:, i] = points[:, i] * 400

colors = ['#762a83','#9970ab','#c2a5cf','#e7d4e8','#d9f0d3','#a6dba0','#5aae61','#1b7837',
         '#d73027','#f46d43','#fdae61','#fee090','#e0f3f8','#abd9e9','#74add1','#4575b4']
events_vertical = []
for i in range(16):
    events_vertical.append({'points': points[4*i:4*i+4, :].tolist(), 'color': colors[i]})
    

for i in range(2):
    points[:, i] = points[:, i] - np.min(points[:, i])
    points[:, i] = points[:, i] / np.max(points[:, i])
    points[:, i] = points[:, i] * 2
    points[:, i] = points[:, i] - 1

with open('../synthetic/static/data/events_vertical_line.json', 'w+') as f:
    json.dump(events_vertical, f)
```

```python
rotate_matrix = np.array([[np.cos(np.pi/4), -np.sin(np.pi/4)],
                         [np.sin(np.pi/4), np.cos(np.pi/4)]])

rotated_points = points @ rotate_matrix
for i in range(2):
    rotated_points[:, i] = rotated_points[:, i] + 1
    rotated_points[:, i] = rotated_points[:, i] * 0.5
    rotated_points[:, i] = rotated_points[:, i] * 400
    
events_rotated = []
for i in range(int(rotated_points.shape[0]/4)):
    events_rotated.append({'points': rotated_points[4*i:4*i+4, :].tolist(), 'color': colors[i]})
    
with open('../synthetic/static/data/events_rotated_line.json', 'w+') as f:
    json.dump(events_rotated, f)
    
```

```python
#creating the events
points = []
for i in range(2):
    for j in range(4):
        x0 = 1 + 12 + 4 * i
        y0 = 1 + 4 * j
        x1 = x0 + 2
        y1 = y0 + 6
        points += [[x0, y0],
                 [x1, y0],
                 [x1, y1],
                 [x0, y1]]

for i in range(4):
    x0 = 1 - 4 + 4 * i
    y0 = 1 + 16 
    x1 = x0 + 6
    y1 = y0 + 2
    points += [[x0, y0],
               [x1, y0],
               [x1, y1],
               [x0, y1]]
    
for i in range(4):
    x0 = 1 + 16 + 4 * i
    y0 = 1 
    x1 = x0 + 6
    y1 = y0 + 2
    points += [[x0, y0],
               [x1, y0],
               [x1, y1],
               [x0, y1]]
    
points = np.array(points).astype(np.float32)
for i in range(2):
    points[:, i] = points[:, i] - np.min(points[:, i])
    points[:, i] = points[:, i] / np.max(points[:, i])
    points[:, i] = points[:, i] * 400

colors = ['#762a83','#9970ab','#c2a5cf','#e7d4e8','#d9f0d3','#a6dba0','#5aae61','#1b7837',
         '#d73027','#f46d43','#fdae61','#fee090','#e0f3f8','#abd9e9','#74add1','#4575b4']
events_two_orientation = []
for i in range(int(points.shape[0]/4)):
    events_two_orientation.append({'points': points[4*i:4*i+4, :].tolist(), 'color': colors[i]})

with open('../synthetic/static/data/events_two_orientation.json', 'w+') as f:
    json.dump(events_two_orientation, f)
```

```python
#creating the events
points = []
for i in range(6):
    x0 = 1 + 4 * i
    y0 = 1 + 12
    x1 = x0 + 6
    y1 = y0 + 2
    points += [[x0, y0],
             [x1, y0],
             [x1, y1],
             [x0, y1]]

for i in range(6):
    x0 = 1 + 12
    y0 = 1 + 4 * i
    x1 = x0 + 2
    y1 = y0 + 6
    points += [[x0, y0],
             [x1, y0],
             [x1, y1],
             [x0, y1]]
    
points = np.array(points).astype(np.float32)

for i in range(2):
    points[:, i] = points[:, i] - np.min(points[:, i])
    points[:, i] = points[:, i] / np.max(points[:, i])
    points[:, i] = points[:, i] * 2
    points[:, i] = points[:, i] - 1
    
rotate_matrix = np.array([[np.cos(np.pi/4), -np.sin(np.pi/4)],
                         [np.sin(np.pi/4), np.cos(np.pi/4)]])

points = points @ rotate_matrix
for i in range(2):
    points[:, i] = points[:, i] + 1
    points[:, i] = points[:, i] * 0.5
    points[:, i] = points[:, i] * 400
    

colors = ['#9970ab','#c2a5cf','#e7d4e8','#d9f0d3','#a6dba0','#5aae61',
         '#f46d43','#fdae61','#fee090','#e0f3f8','#abd9e9','#74add1']
events_cross = []
for i in range(int(points.shape[0]/4)):
    events_cross.append({'points': points[4*i:4*i+4, :].tolist(), 'color': colors[i % len(colors)]})
    
with open('../synthetic/static/data/events_cross.json', 'w+') as f:
    json.dump(events_cross, f)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (8, 8))
for i, e in enumerate(events_vertical):
    x0 = e['points'][0][0]
    y0 = e['points'][0][1]
    x1 = e['points'][1][0]
    y1 = e['points'][2][1]
    rect = patches.Rectangle((x0, y0), x1 - x0, y1 - y0, linewidth=1, 
                             edgecolor= "#303030", facecolor=e['color'], alpha = 0.5)
    ax[0][0].add_patch(rect)
    

ax[0][0].set_title("Vertical sequences")
ax[0][0].plot()

for i, e in enumerate(events_rotated):
    rect = patches.Polygon(e['points'], linewidth=1, 
                             edgecolor= "#303030", facecolor=e['color'], alpha = 0.5)
    ax[0][1].add_patch(rect)

ax[0][1].plot()
ax[0][1].set_xlim((0, 400))
ax[0][1].set_ylim((0, 400))
ax[0][1].set_title("Diagonal sequences")

for i, e in enumerate(events_two_orientation):
    x0 = e['points'][0][0]
    y0 = e['points'][0][1]
    x1 = e['points'][1][0]
    y1 = e['points'][2][1]
    rect = patches.Rectangle((x0, y0), x1 - x0, y1 - y0, linewidth=1, 
                             edgecolor= "#303030", facecolor=e['color'], alpha = 0.5)
    ax[1][0].add_patch(rect)
ax[1][0].set_title("Sequence forming 'L'")
ax[1][0].plot()

for i, e in enumerate(events_cross):
    rect = patches.Polygon(e['points'], linewidth=1, 
                             edgecolor= "#303030", facecolor=e['color'], alpha = 0.5)
    ax[1][1].add_patch(rect)
ax[1][1].set_xlim((0, 400))
ax[1][1].set_ylim((0, 400))
ax[1][1].set_title("Sequence forming 'X'")

plt.tight_layout()

plt.savefig('figs/generated-datasets.svg')
plt.show()
```
